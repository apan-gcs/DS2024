```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example for plotting the following x-y coordinates
x = [0, 1, 2,  3,  4]
y = [1, 2, 5, 10, 17]
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
# Creating figure and plotting sin(x) from [0, 4Ï€]
fig, (ax1, ax2) = plt.subplots(2, 1)  # Create a fig/axes with two rows
x = np.linspace(0, 4*np.pi, 1000)     # Generate 1000 evenly spaced points from 0 to 4
y = np.sin(x)                         # Generate the y-values for sin(x)
ax1.plot(x, y, linestyle='--', color='crimson')
```

```python
# Plot the 9 major points (peaks, troughs, midline)
x_points = np.linspace(0, 4*np.pi, 9)
y_points = np.sin(x_points)
ax1.scatter(x_points, y_points, color='black', zorder=2)
fig  # Show the plot
```

```python
# Edit titles and labels
fig.suptitle("AP - My First Plot")
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
fig
```

```python
# Plotting cos(x) with points on the second axis
# x = np.linspace(0, 4*np.pi, 1000)    # x is already defined above, can be reused
y2 = np.cos(x)                         # Generate the y-values for cos(x)
ax2.plot(x, y2, linestyle='--', color='teal')

x_points = np.linspace(0, 4 * np.pi, 9)
y_points = np.cos(x_points)
ax2.scatter(x_points, y_points, color='black', zorder=2)

# Edit titles and labels
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')

fig
```

```python
# Finally, set tight layout auto-format and to prevent x-axis label from overlapping
# Right-click and Copy or Save as...
fig.tight_layout()
fig
```
